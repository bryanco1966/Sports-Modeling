```python
#eff ratio ratio by line
nba['eff_ratio'] = ((nba.mov_5_deff )/(nba.mov_5_away_def_eff ))  
nba['eff_ratio2'] = (nba.mov_5_oeff/ nba.mov_5_away_off_eff) 
 
```

```python
nba['free_throw_ratio'] = nba.mov_5_fta/nba.mov_5_away_fta
nba.free_throw_ratio.dtype
```

```python
nba['rebound_ratio'] = nba.mov_5_tot/nba.mov_5_away_total_reb
```

```python
nba['score_ratio'] = nba.mov_5_home_score/nba.mov_5_away_score
```
